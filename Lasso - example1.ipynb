```python
import pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.metrics import accuracy_score
```

```python
x,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, random_state=13, noise=15)
plt.scatter(x,y);
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2)
```

```python
l= LinearRegression()
l.fit(xtrain, ytrain)
```

```python
ypred = l.predict(xtest)
```

```python
l.coef_
```

```python
l.intercept_
```

```python
alpha = [0,1,5,10,30]
plt.figure(figsize=(15,5))
plt.scatter(x,y)
for i in alpha:
    l=Lasso(alpha=i)
    l.fit(xtrain, ytrain)
    plt.plot(xtest, l.predict(xtest),label=f'alpha{i}')
plt.legend()
plt.show();
```
