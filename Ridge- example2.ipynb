```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = make_regression(n_samples=500, n_features=1, n_informative=1, n_targets=1, noise=15, random_state=13)
plt.figure(figsize=(15,5))
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
l= LinearRegression()
l.fit(x,y)
print(l.coef_)
print(l.intercept_)
```

```python
from sklearn.linear_model import Ridge
r1=Ridge(alpha=0)
r1.fit(x,y)
print(r1.coef_)
print(r1.intercept_)
```

```python
r2=Ridge(alpha=10)
r2.fit(x,y)
print(r2.coef_)
print(r2.intercept_)
```

```python
r3=Ridge(alpha=100)
r3.fit(x,y)
print(r3.coef_)
print(r3.intercept_)
```

```python
plt.plot(x,y,'b.');
plt.plot(x,l.predict(x), color='red',label = 'Linear model')
# plt.plot(x,r1.predict(x), color='black',label = 'alpha =0')
plt.plot(x,r2.predict(x), color='green',label = 'alpha =10')
plt.plot(x,r3.predict(x), color='yellow',label = 'alpha =100')
plt.legend()
plt.show()
```
